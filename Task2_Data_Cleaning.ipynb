```python
import pandas as pd

import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv("Telco_Customer_Churn_Dataset.csv")

print("_" * 60)
print("ORIGINAL DATASET")
print("_" * 60)
print("Shape:", df.shape)
```

```python
df.columns = (
    df.columns
    .str.strip()
    .str.lower()
    .str.replace(" ", "_")
)
```

```python
print("\nColumn Names Standardized Successfully!")
```

```python
print("\nMissing Values Before Cleaning")
print(df.isnull().sum())
```

```python
df["totalcharges"] = pd.to_numeric(df["totalcharges"], errors="coerce")
```

```python
df["totalcharges"].fillna(df["totalcharges"].median(), inplace=True)

print("\nMissing Values After Filling")
print(df.isnull().sum())
```

```python
duplicates = df.duplicated().sum()
print("\nDuplicate Rows Before:", duplicates)
```

```python
df.drop_duplicates(inplace=True)

print("Duplicate Rows After:", df.duplicated().sum())
```

```python
print("\nData Types")
print(df.dtypes)
```

```python
print("\nDataset Information")
df.info()
```

```python
print("\nFinal Dataset Shape:", df.shape)
```

```python
df.to_csv("Cleaned_Telco_Customer_Churn.csv", index=False)

print("\n" + "_" * 60)
print("DATA CLEANING COMPLETED SUCCESSFULLY")
print("Cleaned dataset saved as:")
print("Cleaned_Telco_Customer_Churn.csv")
print("_" * 60)
```
